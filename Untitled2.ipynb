```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Paramtres de l'hlice
radius = 1.0  # Rayon du cylindre
height = 5.0  # Hauteur du cylindre
num_points = 1000  # Nombre de points sur l'hlice
num_turns = 5  # Nombre de tours complets

# Gnrer les coordonnes de l'hlice
theta = np.linspace(0, 2 * np.pi * num_turns, num_points)
x = radius * np.cos(theta)
y = radius * np.sin(theta)
z = np.linspace(0, height, num_points)

# Crer une figure 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Tracer l'hlice
ax.plot(x, y, z)

# Dfinir les limites des axes
ax.set_xlim(-radius, radius)
ax.set_ylim(-radius, radius)
ax.set_zlim(0, height)

# Afficher la figure
plt.show()
```
